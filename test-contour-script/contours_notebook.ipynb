```python
"""
Source: Publication: https://www.researchgate.net/publication/258443282_Fast_automated_yeast_cell_counting_algorithm_using_bright-field_and_fluorescence_microscopic_images

see: Image processing"""

import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def countours(filepath, treshhold, contour_size, imshow=False):
    """Counts cell number dependent of number of contours(bigger than contour_size) in binarasied(dependend of treshhold) photo"""
    
    im = cv2.imread(filepath,) # reads image from filepath
    imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY) # change image to gray
    ret,thresh = cv2.threshold(imgray, treshhold, 255, 0) # make binary image "black/white"
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(imgray, contours, -1, (255,0,0), 3) # finds contours on binary image
    
    if imshow == True: # shows image if imshow=True in parameters
        plt.imshow(imgray)
        
    cell_contours = [len(c) for c in contours if len(c) > contour_size]
    return len(cell_contours) # count contours

countours("14.TIF", 8, 50, imshow=True)
```
